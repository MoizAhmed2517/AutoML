```python
from pycaret.datasets import get_data
from pycaret.regression import *
```

```python
data1 = get_data(dataset='traffic')
```

```python
dataset = setup(data=data1, target='traffic_volume', session_id=438)
```

```python
best = compare_models()
```

```python
# evaluate_model(best)
plot_model(best, plot = 'residuals')
```

```python
tuned_best = tune_model(estimator=best)
```

```python
save_model(tuned_best, 'C:/Users/HP/ML Models/model/traffic_pycaret')
```

```python
model_traffic = load_model('C:/Users/HP/ML Models/model/traffic_pycaret')
```

```python
predictions = predict_model(model_traffic, data=data1)
predictios.head()
```
